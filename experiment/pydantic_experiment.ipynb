```python
#standard python behaviour
def play_game(score: int):
    print(f"Your Score is {score}!")

#We Said 'score' is an integer , but we give it a string .
#Python doesn't care ,it just runs the code without any error .
play_game("One 64478")
```

```python
from pydantic import BaseModel

#we create a pydantic 'model' by inheriting from basemodel

class player(BaseModel):
    name : str
    score : int
```

```python
player1 = player(name = "Sindhura" ,score = 78)
print("player1 :",player1)
```

```python
player2 = player(name = "Hari" ,score = "60")
print("player2 :",player2)
print("------------------------")
```

```python
player3 = player(name = "Alex" ,score = "FiftyFive")
print("player3 :",player3)
print("------------------------")
```

ONE AGENT WITH PERSONA

```python
pirate_agent = Agent(
    name="monkey-d-luffy",
    description="a great caption",
    model='groq:llama-3.3-70b-versatile',
    system_prompt=(
        "You are Captain Monkey D Luffy. Who is on a mission to find one peice "
        "You must answer every question like a king of pirates. "
        "Never drop your character."
    )
)
```

```python
pirate_agent
```

```python
response = pirate_agent.run_sync("Explain how to get food on a ship. Answer in one short sentence.")

print("Pirate Agent Answer:", response.output)
```
